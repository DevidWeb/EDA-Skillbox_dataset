```python
# Для анализа данных в рамках данной курсовой работы нам понадобится рад библиотек.
# Загрузим библиотеки:
import numpy as np
import pandas as pd
import missingno as msno
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator
```

```python
# С помощью методов библиотеки pandas загрузим предоставленные данные в notebook курсовой работы
students_df = pd.read_csv('students.csv')
students_df.pop('id_') #Уберем неизвестные данные
courses_df = pd.read_csv('courses.csv')
courses_contents_df = pd.read_csv('course_contents.csv')
progresses_df = pd.read_csv('progresses.csv')
progress_phases_df = pd.read_csv('progress_phases.csv')#, parse_dates=['start_date', 'finish_date'])
```

```python
# Объеденим данные в один общий датасет
# Для объединения необходимо, чтобы названия столбцов, по которым ведется объединение данных, совпадали.
# Объединим данные из файлов 'progresses.cvs' и 'progress_phases_df.csv', предварительно переименовав колонку 'id'
progresses_df = progresses_df.rename (columns= {'id': 'progress_id'})
data_df = progress_phases_df.merge(progresses_df,  on = 'progress_id', how = 'left')
# Объединим полученные данные с данными из файла 'courses_df', предварительно переименовав колонку 'id'
courses_df = courses_df.rename (columns= {'id': 'course_id'})
data_df = data_df.merge(courses_df,  on = 'course_id', how = 'left')
#Удалим неизвестную колонку
data_df.pop('Unnamed: 0')
# Объединим полученные данные с данными из файла 'courses_contents_df.cvs'
# В данном случае данные "пересекаются" по нескольким столбцам, по ним и будем объединять данные
data_df = data_df.merge(courses_contents_df,  on = ['course_id', 'module_number', 'lesson_number'], how = 'inner')
```

```python
#Посчитаем количество модулей на каждом курсе.
modules_quantity = data_df.groupby('title')['module_title'].nunique().reset_index()
modules_quantity.rename(columns= {'module_title': 'module_title_count'}, inplace=True)
print('Таблица 1. Количество модулей в курсах')
modules_quantity.sort_values(by='module_title_count')
```

```python
#Посчитаем количество уроков в каждом модуле на каждом курсе.
#Для этого сгруппируем датасет по курсам  и модулям
lessons_quantity = data_df.groupby(['title', 'module_title'])['lesson_title'].nunique().reset_index()
lessons_quantity.rename(columns= {'lesson_title': 'lesson_title_count'}, inplace=True)
print('Таблица 2. Количество уроков в каждом модуле на курсах')
lessons_quantity.sort_values(by='title')
```

```python
#Посчитаем количество студентов  на каждом курсе.
#Для этого сгруппируем датасет по курсам
student_quantity = data_df.groupby('title')['student_id'].nunique().reset_index()
student_quantity.rename(columns= {'student_id': 'student_count'}, inplace=True)
print('Таблица 3. Количество студентов на курсе')
student_quantity.sort_values(by='student_count')
```

```python
student_quantity['student_count'].sum()
```

```python
#Посчитаем минимальный, максимальный, средний, медианный возраст студентов
#Для этого предварительно рассчитаем возрас каждого студента опираясь на дату рождения.
#Так как ряд дат отсутствует, а некоторые данные введены с ошибкой (данные нельзя преобразовать в тип Дата), воспользуемся собственной функцией:

def calculate_age(birthday):
    """
    Функция для расчета возраста
    Args:
        birthday (str): Дата дня рождения
    Attribute:
        today (class?): Сегодняшняя дата
    return:
        age (int): Возраст
    """
    today = pd.Timestamp('now')
    try:
        birthday = pd.to_datetime(birthday)
        age = today.year - birthday.year
        if today.month > birthday.month or (birthday.month == today.month and today.day > birthday.day):
            age += 1
        if age <= 10 or age > 90: #Добавим граничные условия по возрасту от 10 и до 90 лет. Но в данном случае я руководствуюсь собственным опытом (по возможности проходить учебу). Возможно хозяин бизнеса видит это по-другом, например, возраст окончания школы (17) и выхода на пенсию(65).
            return None
        return int(age)
    except:
        return None

#Создадим новую колонку с возрастом студентов, рассчитанным с помощью функции "calculate_age"
students_df['age'] = students_df['birthday'].apply(calculate_age)
```

```python
print('Таблица 4. Статистические значения возраста всех студентов')
students_df["age"].describe()
```

```python
#Для расчета статистический значений возраста по курсам, необходимо объединить датасет data_df со students_df
#Предварительно переименуем колонки
students_df = students_df.rename (columns= {'id': 'student_id'})
#Создадим датафрейм состоящий из части базового датасета и students_df
data_df2 = data_df[['title', 'student_id']].merge(students_df[['student_id', 'age']], on = 'student_id')
#уберем повторяющихся студентов путем группировки
data_df2 = data_df2.groupby(['title', 'student_id']).agg({'age': 'first'})
#расчет статистических значений
data_df2 = data_df2.groupby('title')['age'].agg(['mean', 'median', 'max', 'min'])
```

```python
print('Таблица 4. Статистические значения возраста студентов на каждом курсе')
data_df2
```

```python
#Построим bar-chart, отражающий количество студентов на каждом курсе.
#для лучшего визуального восприятия применим сортировку по количеству студентов от большего к меньшему
student_quantity = student_quantity.sort_values(by='student_count', ascending=False)

#"Начертим" наш bar
#Создадим окно графика:
plt.figure(figsize=(25,10))

#Создадим столбчатую диаграмму:
plt.bar(student_quantity['title'],  student_quantity['student_count'])

#Настроим внешний вид диаграммы:
plt.title('Количество студентов на курсе', fontsize=20)
plt.xlabel('Курс', fontsize=16)
plt.ylabel('Количеств студентов', fontsize=16)
plt.xticks(rotation = 45)
plt.show()
```

```python
#Построим bar-chart, отражающий количество студентов на каждом курсе и в сфере.
#Для этого добавим в наша датасет информацию по сфере курса:
student_quantity = student_quantity.merge(data_df[['title', 'field']], on = 'title').groupby('title').agg({'field': 'first','student_count':'first'}).sort_values(by='student_count')

#Создадим окно графика:
fig, ax = plt.subplots(figsize = (20,10))

#Создадим горизонтальную столбчатую диаграмму зависимости количество студентов от курса и сферы:
sns.barplot(ax = ax, data=student_quantity, x='student_count', y=student_quantity.index, ci=None, alpha = 0.1, hue = 'field', dodge = False)

#Добавим линию медианы на диаграмму
student_median_quantity = student_quantity['student_count'].median()
ax.axvline(x=student_median_quantity, color = 'green', linewidth = 1, linestyle = '--', label = 'Медиана количества студентов')

#Настроим внешний вид диаграммы:
ax.set_title('Диаграмма зависимости количества студентов от сферы и курса', fontsize = 20)
ax.set_xlabel('Количество студентов на курсе', fontsize = 16)
ax.set_ylabel('Название курса', fontsize = 16)


sns.despine(left=True, bottom=True, right=True)
plt.setp(ax.get_xticklabels(), fontsize = 14)
plt.setp(ax.get_yticklabels(), fontsize = 14)
ax.xaxis.set_major_locator(MultipleLocator(250))

plt.legend(fontsize=14, title='Сфера курса', title_fontsize = 16, loc = 'upper right')


plt.show()
```

```python
#Для того, чтобы учесть все месяцы, потребавшиеся на выполнения домашнего задания, напишем следующую функцию:
def calculate_period(start_finish_df):
    """
    Функция для составления списка дат-месяцев в заданном периоде
    Args:
        start_finish_df (list): Хранит в себе две даты: начало и конец периода

    return:
        period (list): Список с дат-месяцами
        start_finish_df[0]: Дата начала периода
    """
    try:
        period = pd.date_range(start=start_finish_df[0], end=start_finish_df[1], freq = 'MS', closed='right')
        if len(period) > 0:
            return list(period)
        else:
            return start_finish_df[0]
    except:
       return start_finish_df[0]

#Результаты работы функции поместим в отдельный столбец нашего датафрейма
data_df5['period'] = data_df5.loc[:,['start_date', 'finish_date']].apply(calculate_period, axis=1)
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).
2. **(m)** Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы
3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
#Формируем новый датафрейм для выявления сезонности для каждого курса
#Отбираем данные в которых есть Домашняя работа
data_df8_1 = data_df[data_df['is_homework'] == True][['title','module_number', 'module_title','lesson_number','student_id' , 'status', 'field']]

#Сделаем два датафрейма
#1) Первый датафейм отвечает за последнее сделанное дз в модуле. Для этого мы выбираем данные в которых 'status' == 'done'
# группируем по крусу, модулю и уроку, а затем выбираем данные согласно последнему уроку
data_df8_2 = data_df8_1[data_df8_1['status'] == 'done']
data_df8_2 = data_df8_2.groupby(['title','field', 'module_number', 'module_title', 'lesson_number'])['student_id'].count().reset_index()
data_df8_2 = data_df8_2.groupby(['title', 'field','module_number', 'module_title']).agg({'lesson_number':'last','student_id':'last'}).reset_index()
data_df8_2 = data_df8_2.rename(columns= {'student_id': 'student_done'})
#2) Второй датафейм отвечает за первое дз, к которому приступили в модуле. Для этого мы выбираем данные в которых 'status' == 'done' и 'start'
# группируем по крусу, модулю и уроку, а затем выбираем данные согласно первому урок
#data_df8_3 = data_df8_1[(data_df8_1['status'] == 'done') | (data_df8_1['status'] == 'start')]
data_df8_3 = data_df8_1.groupby(['title','field', 'module_number', 'module_title', 'lesson_number'])['student_id'].count().reset_index()
data_df8_3 = data_df8_3.groupby(['title','field', 'module_number', 'module_title']).agg({'lesson_number':'first','student_id':'first'}).reset_index()
data_df8_3 = data_df8_3.rename(columns= {'student_id': 'student_start'})
```

```python
#Соединим два датафрейма в один
data_df8_4 = data_df8_2.merge(data_df8_3,  on = ['title', 'field','module_number','module_title'], how = 'left')
data_df8_4
```

```python
#посчитаем конверсию на основе объединенного датафрейма
#для этого с помощью цикла разобьем датафрейм по курсам и для каждого курса посчитаем конверсию

help_list = list()
for i in data_df8_4['title'].unique():
    #Создаем отдельный датасет курса
    data_df_title = data_df8_4[data_df8_4['title'] == i][['title', 'field', 'module_number', 'module_title', 'student_done', 'student_start']]

    #добавляем столбец с изменением количества студентов на курсе
    data_df_title['conversion'] = data_df_title['student_start'] / data_df_title['student_done'].shift() * 100
    #добавляем полученный датасет в список
    help_list.append(data_df_title)

#Соединяем все полученные датасеты в один.
data_df8_5= pd.concat(help_list)
data_df8_5
```

```python
fig, ax = plt.subplots(figsize = (20,10))

sns.barplot(data = data_df8_5, x='module_number', y='conversion', ax=ax, ci =None, hue='title')

fig.suptitle('Конверсия по курсам', y=0.92, color='grey', weight='bold', fontsize = 20 )
ax.set_xlabel('Номер модуля', fontsize = 12)
ax.set_ylabel('Конверсия, %', fontsize = 12)
ax.grid(axis = 'y', color='gray', linestyle = ':',  linewidth=1)
sns.despine(right=True)
plt.legend(fontsize=12, title='Названия курсов', title_fontsize = 16, loc = 'upper right')
plt.show()
```

```python
#На основе датфрейма построим 15 графиков для каждого курса

#В этот раз построим  графики в виде таблицы 5х3
fig, axs = plt.subplots(nrows=5, ncols=3, sharey=False, sharex=False, figsize=(30,30))
plt.subplots_adjust(wspace=0.3, hspace=0.5)

#Словарь цветов
field_color = {}
for i, field_ in enumerate(data_df8_5['field'].unique()):
    field_color[field_] = color_[i]

help_list = list()
#Построение графиков и настройка внешнего вида:
j = 0
for i, course in enumerate(data_df8_5['title'].unique()):
    data = data_df8_5[(data_df8_5['title'] == course)][['field','module_number', 'conversion', 'student_start', 'student_done']]
    sns.barplot(data=data, y='module_number', x='conversion', ax=axs[i//3,j], ci=None, orient = 'h',alpha = 0.1, color=field_color[data['field'].unique()[0]], label = f"Сфера курса:{data['field'].unique()[0]}")

    #Дополнительно нанесем на график общую конверсию курса (отношение количества студентов завершивших последнее ДЗ на курсе к количеству студентов приступивших к выполнению первого дз на курсе)
    total_conversion = data[data['module_number'] == data['module_number'].max()]['student_done'].values / data[data['module_number'] == data['module_number'].min()]['student_start'].values * 100
    axs[i//3,j].axvline(x=total_conversion, color = 'red', linewidth = 1, linestyle = 'solid', label = 'Общая конверсия курса')
    help_list.append(total_conversion)

    axs[i//3,j].grid(axis = 'x', color='gray', linestyle = ':',  linewidth=1)
    axs[i//3,j].set_title(f'Конверсия по курсу: {course}', pad=10, weight='bold')
    axs[i//3,j].set_xlabel('Конверсия, %', color='navy')
    axs[i//3,j].set_ylabel('Номер модуля', color='navy')
    axs[i//3,j].axvline(x=data['conversion'].median(), color = 'green', linewidth = 1, linestyle = '--', label = 'Медиана конверсии курса')
    axs[i//3,j].legend(loc = 'upper left')

    sns.despine(left=True, bottom=True, right=True)
    if data['conversion'].max() <= 150:
        axs[i//3,j].set_xticks([i for i in range(0, int(data['conversion'].max())+10, 10)])
    else:
        axs[i//3,j].set_xticks([i for i in range(0, int(data['conversion'].max()+25), 25)])

    j=j+1
    if j == 3:
        j=0

fig.suptitle('Конверсия по курсам', y=0.92, color='grey', weight='bold', fontsize = 20 )

plt.show()
```
